# Import packages for plotting graphs and manipulating data

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Plot the noisy data (x,y1) and clean data (x,y2)

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (clean) y-coordinate data 
# y2 : (noisy) y-coordinate data

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 			# clean points
y2      = y1 + nn * std		# noisy points

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

# Define a function that plots polynomial curve which fits the noisy data by the least square error with regularization.

```python
def withregularization(p,lamb):
    A=np.zeros((num+p, p), dtype=float)
    
    for i in range(num,num+p):
        A[i,i-num]=math.sqrt(lamb)
    for i in range(num):
        count=0
        for j in range(p):
            A[i,j]=x[i]**count
            count=count+1;
    
    yy=np.zeros(num+p,dtype=float)
    for i in range(num):
        yy[i]=y2[i]
        
    theta=((np.linalg.pinv(A.T.dot(A))).dot(A.T)).dot(yy)
    
    y_=np.zeros(num,dtype=float)
    
    for i in range(num):
        for j in range(p):
            y_[i]+=theta[j]*(x[i]**j)
    
    print('     with lambda: ',lamb)
    plt.plot(x,y2,'k.',x,y_,'r')
    plt.show()
```

# Try with varying the p  and the regularization parameter.

```python
print('when p: ',6)
for i in range(5,18):
    withregularization(6,2**(i))
print('--------------------------------------')

print('when p: ',7)
for i in range(5,18):
    withregularization(7,2**(i))
print('--------------------------------------')

print('when p: ',8)
for i in range(5,18):
    withregularization(8,2**(i))
print('--------------------------------------')

print('when p: ',9)
for i in range(5,18):
    withregularization(9,2**(i))
print('--------------------------------------')
```
